```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import matplotlib.font_manager as fm
import matplotlib as mpl 
import numpy as np


path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf' # 나눔 고딕
font_name = fm.FontProperties(fname=path, size=20).get_name() # 기본 폰트 사이즈 : 20
plt.rc('font', family=font_name)
fm._rebuild()
mpl.rcParams['axes.unicode_minus'] = False
%matplotlib inline
```

```python
#bar 그래프
#x나 y중 하나의 값이 카테고리형일 경우 bar 형태 차트가 효과적인 시각화 방법이다
#가로 방향으로 그리려면 barh 로 그릴 수 있습니다
#plt.bar(x,y,color='색깔',width='폭')

x = np.arange(4)
y = [100,90,78,90]
xlab = ['국어','수학','역사','사회']

plt.bar(x,y) # x 축은 [0,1,2,3]로 y 축은 [100,90,78,90]으로 기준을 하기 때문에 1 ~ 100 까지 20 견격이라고 볼 수 있다
plt.xticks(x,xlab)  # [0,1,2,3] , ['국어','수학','역사','사회']  xticks는 이제 ticks를 정하는 건데 간격은 0,1,2,3 으로 하는데 표시를 '국어','수학','역사','사회' 로 나타내기 위함이다
plt.ylabel('점수')
plt.show()
x
```

```python
x = np.arange(4)
y = [100,90,80,60]
xlabel = ['국어','수학','영어','역사']

plt.figure(figsize=(10,5))

ex1 = plt.subplot(1,2,1)
plt.bar(x,y)
plt.xticks(x,xlabel)
plt.ylabel('점수')


ex1 = plt.subplot(1,2,2)
plt.bar(x,y)
plt.xticks(x,xlabel)
plt.ylabel('점수')
```

```python
plt.barh(x, y)
plt.yticks(x, xlabel) # 이게 barh는 x축에 위치가 y축에 있고 y축의 위치가 x 축에 있다 하지만 xticks는 그걸 모르기 때문에 x 축에 tick 을 정하고 싶다면 ytics로 해줘야 한다
plt.title('안녕 barh')
plt.show()
```

```python

plt.figure(figsize=(10,5))

ex1 = plt.subplot(1,2,1)
plt.barh(x,y, label='점수')
plt.legend(loc='upper right')
plt.yticks(x,xlabel)
plt.title('첫번째 그래프')

ex2 = plt.subplot(1,2,2)
plt.bar(x,y, label='점수')
plt.xticks(x,xlabel)
plt.legend(loc='upper right')
plt.title('두번째 그래프')
```

```python

x = np.arange(5)
women = [80, 70, 40, 90, 60]
men = [70, 80, 45, 75, 80]
xlabel = ['국어', '수학', '영어', '과학', '미술', '체육']

bar_width = 0.35                                           
plt.bar(x, women, width=bar_width, label='여자')            
plt.bar(x + bar_width, men, width=bar_width, label='남자')  
plt.xticks(x + bar_width/2, xlabel)                         
plt.legend(loc='upper right')
plt.show()
```

```python
# 나란한 형태의 그래프를 그림

x = np.arange(5)
jung = [100,60,90,80,70]
sang = [90,95,62,84,67]
hong = [60,50,85,60,90]

xlabel = ['국어','사회','과학','영어','수학']

plt.bar(x,jung,width=0.5)
plt.show()


plt.bar(x,jung, width=0.25, label='정예준')
plt.bar(x+0.25,sang,width=0.25, label='임상운')
plt.bar(x+0.5,hong,width=0.25,label='홍석훈' )
plt.xticks(x +0.25,xlabel) # 중간을 점프한 값에다가 해줘야 가운데에 보인다

plt.legend(loc='upper right')
plt.show()
```

```python
#bar 여러개 그리기
# bottom 옵션을 사용해 겹쳐진 바 차트도 만들 수 있다

plt.bar(x,sang, width=0.25, label='임상운', bottom=jung)
plt.bar(x,jung,width=0.25, label='정예준')
plt.xticks(x,xlabel)
plt.ylabel('점수')
plt.legend(loc='upper right')
plt.show()
```

```python
# bottom='아래에다가 두려고 하는 데이터의 값 변수명을 적어주면 된다'

plt.bar(x,hong, width=0.25, label='홍석훈', bottom=sang)
plt.bar(x,sang, width=0.25, label='임상운')
plt.legend(loc='upper right')
plt.xticks(x,xlabel)

plt.show()
```

```python
# pie 란 
# 그 원으로 되어 있는 그림 맞습니다
# 상대적인 값을 비교할 경우는 pie 차특 선호된다
# plt.pie(카테고리 값, colors=['색상], width='폭', explode='돌출 옵션')

labels = ['데브옵스','데이터 관리자','백엔드','프론트엔드','서버 개발자']
sizes = [20,5,40,25,10] # 100%를 기준으로 알려주면 된다
colors = ['blue','green','red','lightgray','yellow']
```

```python
plt.figure(figsize=(5,5))


ex1 = plt.subplot(1,2,1)
plt.pie(sizes, colors=['blue','green','red','lightgray','yellow'], labels=labels)
plt.title('프로젝트 기여 %')


ex1 = plt.subplot(1,3,3)
plt.pie(sizes, colors=['blue','lightskyblue','green','lightgray','yellowgreen'], labels=labels)
plt.title('프로젝트 기여 %')

```

```python

#labels = ['데브옵스','데이터 관리자','백엔드','프론트엔드','서버 개발자']
#sizes = [20,5,40,25,10] # 100%를 기준으로 알려주면 된다
#colors = ['blue','green','red','lightgray','yellow']

plt.pie(sizes,colors=colors,labels=labels)
plt.title('프로젝트 기여')
plt.show()

plt.pie(sizes,colors=colors,labels=labels)
plt.show()

explode = (0,0.1,0,0,0)  # 0.1 크기만큼을 돌출 시키겠다는 뜻이다

plt.pie(sizes,colors=colors,labels=labels,explode=explode)
plt.title('explode 사용해서 가독성 높이기')
plt.show()


plt.pie(sizes,colors=colors,labels=labels,explode=explode,shadow=True,autopct='%0.1f%%')  #shadow=True 같은 경우에 그래프의 그림자를 만들고
# autopct는 %를 자동으로 계산해서 보여주는 기능으로 0.1f는 즉 소수점 첫째자리까지 %를 만든다는 뜻이다
plt.title('shadow와 autopct 를 사용한 결과')
plt.show()
```

```python
# 빈도 분포가 어떻게 되어 있는지 확인하기 위해서는 hist 차트가 선호됩니다 (횟수를 나타내는 값)
# plt.hist('발생 빈도가 있는 값')

data = np.random.randn(2000)

plt.hist(data)  # hist 그래프란 x 축의 숫자들 사이에 포함되는 값들이 가지고 있는 갯수를 나타내 주는 것이다
plt.show()

plt.hist(data, label='100')  
plt.hist(data,bins=50, label='50')#bin이란 것을 늘리면 x에 써져있는 구간들을 더 나눠서 그 사이에 숫자들의 갯수를 더 정밀하게 알 수 있다
#위에 그래프를 보면 알 수 있듯 bins로 범위를 늘리면 그 갯수가 분산되므로 당연히 적어갯수를 보이는 것처럼 나온다 하지만 
#만약 저 범위를 원래대로 되돌리면 다시 많은 갯수를 가진 값이란 것을 알 수 있다
plt.legend(loc='upper right')
plt.show()


#누적 분포
plt.hist(data,bins=100,cumulative=True) # 누적 분포 함수를 알 수 있다고 한다 근데 나는 모르겠다고 한다
plt.show()
```

```python
customer_data={'방문횟수':[2,3,5,7],'클릭수':[15,18,25,35],
               '체류시간':[60,200,200,250],'구매':[0,1,0,1]}
df=pd.DataFrame(customer_data,index=['고객A','고객B','고객C','고객D'])

print(df)

plt.scatter(df['방문횟수'],df['클릭수']) #(2,15) (3,18) 등등
plt.show()


plt.scatter(df['방문횟수'],df['클릭수'],c=df['구매'])
#c 매개변수에 변수를 입력해서 사용할 수 있다
# 즉 df 의 값이 같은 값들끼리 같은 색으로 출력이 된다

plt.scatter(df['방문횟수'],df['클릭수'],c=df['구매'],cmap='Greens')
plt.show()
```
